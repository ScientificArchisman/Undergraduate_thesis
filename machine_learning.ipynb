```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.style.use(['science', 'notebook', 'grid'])
divider = '===================================='
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
from sklearn.linear_model import SGDRegressor
from sklearn.neighbors import RadiusNeighborsRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor, VotingRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.cross_decomposition import PLSRegression
from sklearn.linear_model import LinearRegression, LogisticRegression, SGDRegressor
```

```python
data = pd.read_csv("feature_df.csv")
data.tail()
```

```python
data2 = pd.read_csv("filtered_data.csv").drop(columns = ["Condition"])
data2.head()
```

```python
data = data.sample(frac=0.2)
data2 = data2.sample(frac=0.1)
```

```python
X = data.iloc[:, 1:].values
X2 = data2.iloc[:, :-1].values
y = data.iloc[:, 0].values
y2 = data2.iloc[:, -1].values

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y2 = label_encoder.fit_transform(y2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.5, random_state=0)
```

```python
reg1 = GradientBoostingRegressor()
reg2 = RandomForestRegressor(random_state=1)
reg3 = DecisionTreeRegressor()
reg4 = RadiusNeighborsRegressor(radius = 4)
reg5 = KNeighborsRegressor(n_neighbors=2)
reg6 = LinearRegression()
reg7 = GaussianProcessRegressor()
reg8 = LogisticRegression()
reg9 = SGDRegressor()
reg10 = VotingRegressor([('gb', reg1), ('rf', reg2)]) #('dt', reg3), 
                        #  ('rnr', reg4), ('knr', reg5), ('lr', reg6), 
                        #  ('gpr', reg7), ('sgdr', reg9)])

regressors = [reg1, reg2, reg3, reg4, reg5, reg6, reg7, reg8, reg9]
```

```python
reg10
```

```python
for reg in regressors:
    reg.fit(X_train, y_train)
    y_pred = reg.predict(X_test)
    print(reg)
    print("R2 score: ", r2_score(y_test, y_pred))
    print(divider)
```

```python
for reg in regressors:
    reg.fit(X2_train, y2_train)
    y2_pred = reg.predict(X2_test)
    print(reg)
    print("R2 score: ", r2_score(y2_test, y2_pred))
    print(divider)
```

```python
reg10.fit(X2_train, y2_train)
y2_pred = reg10.predict(X2_test)
print(reg10)
print("R2 score: ", r2_score(y2_test, y2_pred))
print(divider)
```
