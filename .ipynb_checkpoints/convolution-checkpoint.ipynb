# The Smoke Function
We define a **Smoke function**
$$
S(t)
$$
It symbolises how much smoke there is in the room at time $t$ because of 1 match stick, i,e, as a function of time.

## The Match Stick Function

Another Function which we may want to look at is the **Match Stick Function**. This is a function which symbolises how many match sticks are being lit in the room at time $t$.

$$ f(t) $$


## Amount of smoke in the room:
1. At minute 0, $f(0) = 1$  
    **Smoke** = $$1\times S(0)$$
2. At minute 1,$f(1) = 2$  
    **Smoke** = $$2\times S(0) + 1\times S(1)$$
3. At minute 2, $f(2) = 3$  
    **Smoke** = $$3\times S(0) + 2\times S(1) + 1\times S(2)$$

## What do we understand from this?
* The amount of smoke in the room at any arbitrary time frame is just the sum of the contributions from all the match sticks that have been lit at each minute.

* We combine functions such that we can track their interaction with each other through time.

## Generalising this:

$$ f*s = \sum_{i=0}^{n} f(i) \times s(n-i) $$

Obviously, as the number of matchsticks is discrete, this is a discrete sum.   

However, we can generalise this for continuous functions as well.
$$ f*s = \int_{0}^{t} f(\tau) \times s(t-\tau) d\tau $$
