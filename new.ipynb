```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use(["science", "notebook", "grid"])
```

```python
history = pd.read_csv("/Users/archismanchakraborti/Desktop/python files/Dissertation/final_model_params/history.csv")
history = history.drop(columns = ["Unnamed: 0"])
history.head()
```

```python
fig, ax = plt.subplots(3, 2, figsize = (15, 12))
ax1, ax2, ax3, ax4, ax5, ax6 = ax.flatten()

ax1.plot(history["loss"], label = "Training Loss")
ax1.plot(history["val_loss"], label = "Validation Loss")
ax1.set_title("Loss")
ax1.set_xlabel("Epochs")
ax1.set_ylabel("Loss")
ax1.legend()

ax2.plot(history["accuracy"], label = "Training Accuracy")
ax2.plot(history["val_accuracy"], label = "Validation Accuracy")
ax2.set_title("Accuracy")
ax2.set_xlabel("Epochs")
ax2.set_ylabel("Accuracy")
ax2.legend()

ax6.plot(history["precision_1"], label = "Training Precision")
ax6.plot(history["val_precision_1"], label = "Validation Precision")
ax6.set_title("Precision")
ax6.set_xlabel("Epochs")
ax6.set_ylabel("Precision")
ax6.legend()

ax3.plot(history["recall_1"], label = "Training Recall")
ax3.plot(history["val_recall_1"], label = "Validation Recall")
ax3.set_title("Recall")
ax3.set_xlabel("Epochs")
ax3.set_ylabel("Recall")
ax3.legend()

ax4.plot(history["f1_score"], label = "Training F1 Score")
ax4.plot(history["val_f1_score"], label = "Validation F1 Score")
ax4.set_title("F1 Score")
ax4.set_xlabel("Epochs")
ax4.set_ylabel("F1 Score")
ax4.legend()

ax5.plot(history["lr"], label = "Learning Rate")
ax5.set_title("Learning Rate")
ax5.set_xlabel("Epochs")
ax5.set_ylabel("Learning Rate")
ax5.legend()

for ax in ax.flatten():
    ax.locator_params(axis = "both", nbins = 15)

plt.tight_layout()
plt.suptitle("Training and Validation Metrics", fontsize = 20, fontweight = "bold", y = 1.05)
```
